### 1 Задание

Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
# Считаем координаты третьего вектора
# с помощью np.array быстрее можно сделать, но с обычным "списком" нагляднее

xb = [20, 20, 20]
xr = [0, 0, -20]
xg = []
for i in range (len(xb)):
    xg.append(xb[i] + xr[i])
xg
```

```python
# функция расчета длины вектора

def veclen(x):
    vecl = (sum([k**2 for k in x]))**0.5
    print(f'Длина вектора {x} равна {vecl}')    
    return vecl
```

```python
xbl = veclen(xb)
```

```python
xrl = veclen(xr)
```

```python
xgl = veclen(xg)
```

```python
# Проверка результата
xgl**2 + xrl**2
```

### 3. Задание (в программе)

Напишите код на Python, реализующий построение графиков:

окружности,

эллипса,

гиперболы.

```python
POINTS = 1000 # количество  точек
```

```python
R = 2  # Рисуем окружность радиуса R
RSQ = R**2

X0, Y0 = 1, 2 # координаты центра окружности

x = np.linspace((X0 - R), (X0 + R), POINTS)

y1 = [Y0 + (RSQ - (i - X0)**2)**0.5 for i in x]
y2 = [Y0 - (RSQ - (i - X0)**2)**0.5 for i in x]

plt.plot(x, y1);
plt.plot(x, y2);

plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Окружность радиуса R = {str(R)}');
```

```python
A, B = 10, 2  # Рисуем эллипс с полуосями А и В

X0, Y0 = 1, 3 # координаты центра эллипса

x = np.linspace((X0 - A), (X0 + A), POINTS)

y1 = [Y0 + B * (1 - (i - X0)**2 / A**2)**0.5 for i in x]
y2 = [Y0 - B * (1 - (i - X0)**2 / A**2)**0.5 for i in x]

plt.axis([X0 - A - 1, X0 + A + 1 , Y0 - B - 1, Y0 + B + 1])

plt.plot(x, y1);
plt.plot(x, y2);

plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Эллипс с полуосями {str(A)} и {str(B)}');
```

```python
A, B = 5, 2  # Рисуем гиперболу с параметрами А и В

X0, Y0 = 4, 1 # координаты центра фигуры

x = np.linspace((X0 - 3 * A), (X0 - A), int(POINTS / 2))
x2 = np.linspace((X0 + A), (X0 + 3 * A), int(POINTS / 2))

y1 = [Y0 + B * ((i - X0)**2 / A**2 - 1)**0.5 for i in x]
y2 = [Y0 - B * ((i - X0)**2 / A**2 - 1)**0.5 for i in x]
y21 = [Y0 + B * ((i - X0)**2 / A**2 - 1)**0.5 for i in x2]
y22 = [Y0 - B * ((i - X0)**2 / A**2 - 1)**0.5 for i in x2]

plt.axis([X0 - 3 * A - 1, X0 + 3 * A + 1 , Y0 - 3 * B - 1, Y0 + 3 * B + 1])

plt.plot(x, y1);
plt.plot(x, y2);
plt.plot(x2, y21);
plt.plot(x2, y22);


plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Гипербола с параметрами а =  {str(A)} и b = {str(B)}');
```

### 5. Задание (в программе)
1. Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

rcParams['figure.figsize'] = 8, 5

fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5, 0.5)
Y = np.arange(-5,5, 0.5)
X,Y = np.meshgrid(X, Y)
Z = 3*X + 7*Y
Z2 = 3*X + 7*Y + 20 # формула значений оси Z параллельной плоскости
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z2)
ax.scatter(0, 0, 0, 'z', 50, 'red')
ax.scatter(0, 0, 20, 'z', 50, 'green')
show()
```

2. Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
rcParams['figure.figsize'] = 10, 10

x = np.linspace(-2*np.pi, 3*np.pi, 201)

K, A, B, = 2, 3, -4
K2, A2, B2 = 0.5, -2, 3

plt.plot(x, K * np.cos(x - A) + B);
plt.plot(x, K2 * np.cos(x - A2) + B2);

plt.axis([-7, 10, -7, 4])

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show();
```
