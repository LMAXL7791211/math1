ч.1 задача 2. Напишите код, который будет переводить полярные координаты в декартовы.

```python
def polartodecart(R, Alpha):
    x = R * np.cos(Alpha)
    y = R * np.sin(Alpha)
    print(f'Полярные координаты {R, round(np.degrees(Alpha), 3)} переведены в декартовы {round(x, 3), round(y, 3)}')
    return x, y

polartodecart(2, np.pi / 4)  # 45 degrees
polartodecart(1, np.pi / 2)  # 90 degrees
polartodecart(3, np.pi)  # 180 degrees
polartodecart(4, np.pi / 6)  # 30 degrees
polartodecart(4, np.pi / 3)  # 60 degrees
polartodecart(4, -np.pi / 3)  # -60 degrees
```

### 4. Задание (в программе)
1) Решите систему уравнений:


y = x**2 – 1

exp(x) + x·(1 – y) = 1


2) Решите систему уравнений и неравенств:


y = x ** 2 – 1

exp(x) + x·(1 – y) > 1

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-2, 5, 201)
plt.plot(x, 1 - (1 - np.exp(x))/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,25) 
plt.grid(True)
plt.show()

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)

x1, y1 =  fsolve(equations, (-2, 2))
              
print (x1, y1)

x2, y2 =  fsolve(equations, (2, 4))
              
print (x2, y2)

x3, y3 =  fsolve(equations, (4, 15))
              
print (x3, y3)
```

Неравенство exp(x) + x·(1 – y) > 1 преобразовывается в y < 1 - (1 - exp(x)) / x. Т.е. решением неравенства будет область, находящаяся "под" "синим" графиком. Решением системы уравнения и неравентса, соответственно, будет множество значений уравнения-параболы от "первого" корня системы уравнений (-1.582, 1,502) до "второго" (2.618, 5.855) (сами корни не включены), в т.ч. (-1,0), (0, -1), (1, 0), (2, 3) и т.д. ; далее - решением также будут значения параболы от "третьего" корня (не включая корень) до + бесконечности (в т.ч. (5,24), (6, 35) и т.д.
